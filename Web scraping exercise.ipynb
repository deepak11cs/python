```python
import requests
```

```python
import bs4
```

```python
res = requests.get("https://www.thegoldbugs.com/blog")
```

```python
type(res)
```

```python
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
len(soup.select('.sqs-block-content > pre'))
```

```python
text = soup.select('.sqs-block-content > pre')
```

```python
text
```

```python
text = text[0].get_text()
```

```python
text
```

Now we need to extract out every first character just after '-----'

```python
text = text.split('-----')
```

```python
text
```

```python
result = [i[0] for i in text[1:]]
```

```python
result = ''.join(result)
```
